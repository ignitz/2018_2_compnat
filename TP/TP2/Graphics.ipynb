```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import os
import itertools

# DATASET_NAMES=["graph1", "graph2", "graph3"]
DATASET_NAMES=["graph1"]

MACHINES =[
    'paracatu',
    'pomba',
    'velhas',
    'tiete',
    'araguaia',
    'piracicaba',
    'congo',
    'grande',
    'vermelho',
    'claro',
    'doce',
    'urucuia',
    'jordao',
    'madeira',
    'ganges',
    'cipo'
]

ANTS_NUM = [50, 100, 200, 1000]
IT_NUM = [100]
SIGMA = [0.1, 0.3, 0.5, 0.7, 0.9]
ALPHA = [0.1, 0.5, 1.0, 2.0]
BETA = [0.1, 0.5, 1.0, 2.0]
K_ANTS = [0, 1, 5, 10, 15]

def read_file(pathname):
    aux_data = list()
    with open(pathname, 'r') as f:
        while True:
            line = f.readline()
            if len(line) == 0:
                break
            aux_data.append(list(map(float, line.split(';'))))
    return np.matrix(aux_data)

def load_data(
                dataset_name,
                it_num,
                ants_num,
                alpha,
                beta,
                sigma,
                init_phero,
                k_ants,
                hostname
              ):
    origin = 1
    if dataset_name not in DATASET_NAMES:
        raise 'DATASET not available'

    dest = {'graph1': 100, 'graph2': 20, 'graph3': 1000}[dataset_name]

    filename = 'output/' + dataset_name
    filename += '_' + str(it_num) + '_' + str(ants_num)
    filename += '_' + str(origin) + '_' + str(dest) + '_' + str(alpha)
    filename += '_' + str(beta) + '_' + str(sigma) + '_' + str(init_phero)
    filename += '_' + str(k_ants) + '_' + str(hostname)
    filename += '_' + '001.csv'

    data = read_file(filename) if os.path.exists(filename) else None
    return data
```

```python
def do_all_stuff():
    EIXOS = [0, 55, 0, 1.5]

    # ANTS_NUM = [50, 100, 200, 1000]
    # IT_NUM = [100]
    # SIGMA = [0.1, 0.3, 0.5, 0.7, 0.9]
    # ALPHA = [0.1, 0.5, 1.0, 2.0]
    # BETA = [0.1, 0.5, 1.0, 2.0]
    # K_ANTS = [0, 1, 5, 10, 15]

    
    # dataset_name, it_num, ants_num, 
    # alpha, beta, sigma, init_phero, k_ants, hostname

    data = None
    
    all_params = list(
        itertools.product(ANTS_NUM,
                          IT_NUM,
                          SIGMA,
                          ALPHA,
                          BETA,
                          K_ANTS,
                          DATASET_NAMES))

    for ants_num, it_num, sigma, alpha, beta, k_ants, dname in all_params:
        fig, ax = plt.subplots()
        data_col = list()
        for hostname in MACHINES:
            data = load_data(
                dname,
                it_num,
                ants_num,
                alpha,
                beta,
                sigma,
                0.1,
                k_ants,
                hostname
            )
            if data is None:
                continue
            data_col.append(data)

        label = '{}    Iterations: {}     Number of Ants: {}\n$\\alpha$ = {}   $\\beta$ = {}    $\\sigma$ = {}    Elitism: {}'.format(
            dname,
            it_num,
            ants_num,
            alpha,
            beta,
            sigma,
            k_ants
        )
        if len(data_col) == 0:
            continue

        data = data_col.pop(0)
        for x in data_col:
            data += x
        data /= (len(data_col) + 1)
        
        ax.set(xlabel='Iterações', ylabel='Fitness médio', title=label)
        ax.plot(np.arange(0, it_num, 1), data[:,0][:], '-', label=label)
        ax.plot(np.arange(0, it_num, 1), data[:,1][:], '-', label=label)
        ax.plot(np.arange(0, it_num, 1), data[:,2][:], '-', label=label)
        ax.plot(np.arange(0, it_num, 1), [990]*it_num, '-', label=label)

        plt.show()
        fname = '{}_Iterations_{}_NumberOfAnts_{}_Alpha_{}_Beta_{}_Sigma_{}_Elitism_{}'.format(
            dname,
            it_num,
            ants_num,
            alpha,
            beta,
            sigma,
            k_ants
        )
        fig.savefig('img/' + fname + ".pdf")
```

```python
do_all_stuff()
```
